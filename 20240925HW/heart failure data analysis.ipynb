```python
import scipy.stats as stats

Disease = heart_data[heart_data['HeartDisease']== 1]
NoDisease = heart_data[heart_data['HeartDisease']== 0]


new_D = new_data[new_data['HeartDisease']== 1]
new_UD= new_data[new_data['HeartDisease']== 0]
```

```python
print(NoDisease.shape)
print(Disease.shape)
```

```python
print(stats.ttest_ind(Disease['Age'], NoDisease['Age']))

print(stats.ttest_ind(Disease['RestingBP'], NoDisease['RestingBP']))

print(stats.ttest_ind(Disease['MaxHR'], NoDisease['MaxHR']))

print(stats.ttest_ind(new_D['Cholesterol'], new_UD['Cholesterol']))

print(stats.ttest_ind(Disease['Oldpeak'], NoDisease['Oldpeak']))
```

```python
T1 = np.array([
                [len(Disease[Disease['Sex']=="F"]),len(NoDisease[NoDisease['Sex']=="F"])],
                [len(Disease[Disease['Sex']=="M"]),len(NoDisease[NoDisease['Sex']=="M"])]
])
print(T1)
print(stats.chi2_contingency(T1, correction=False))


T2 = np.array([
                [len(Disease[Disease['FastingBS']==1]),len(NoDisease[NoDisease['FastingBS']==1])],
                [len(Disease[Disease['FastingBS']==0]),len(NoDisease[NoDisease['FastingBS']==0])]
])
print(T2)
print(stats.chi2_contingency(T2, correction=False))

T3 = np.array([
                [len(Disease[Disease['ExerciseAngina']=="Y"]),len(NoDisease[NoDisease['ExerciseAngina']=="Y"])],
                [len(Disease[Disease['ExerciseAngina']=="N"]),len(NoDisease[NoDisease['ExerciseAngina']=="N"])]
])
print(T3)
print(stats.chi2_contingency(T2, correction=False))


#chi2, p, dof, ex = stats.chi2_contingency(T, correction=False)
#print(chi2)
#print(dof)
#print(p)
```
